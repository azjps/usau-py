```python
import re
with open("/apps/stonefs1/azhu/libs/usau_fantasy/fantasy_input.txt") as fd:
  text = fd.read().replace("[\xe2\x80\x93]", "__START_USER__").replace("permalinkembedsavereportgive goldreply", "")
  
text = re.sub(r"__START_USER__([0-9a-zA-z_]+) .*", "\"\g<1>\": {", text)
lines = []
for line in text.split("\n"):
  if line.strip() == "":
    continue
  if line.startswith("Men") or line.startswith("Women"):
    gender = line.split(":", 1)[0].split("'", 1)[0]
    assert gender in ('Men', 'Women')
    roster = line.split(":", 1)[1].split(",")
    # if len(roster) != 7:
    #   print "Missing player:", roster
    lines.append("  \"" + gender + "\": [\"" + "\", \"".join([r.strip() for r in roster]) + "\"],")
  elif line.startswith('"'):
    lines.append("  },")
    lines.append(line)
    
print "\n".join(lines)
```

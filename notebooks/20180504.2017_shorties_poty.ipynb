```python
from usau.reports import USAUResults
```

```python
u = USAUResults.from_nationals("d1college", 2017, "Men")
```

```python
def compute_plus_minus(df, g_weight=1, a_weight=1, d_weight=1, turn_weight=-0.5):
  return (g_weight * df["Goals"] +
          a_weight * df["Assists"] +
          d_weight * df["Ds"] +
          turn_weight * df["Turns"])
```

```python
import os
os.getcwd()
u.load_from_csvs(os.path.join(os.getcwd(), '..', 'usau', 'data2'))
```

```python
roster = u.rosters
```

```python
roster["+/-"] = compute_plus_minus(roster)
```

```python
roster.loc[roster.Team == "Colorado State"]
```

```python
(roster.sort_values("+/-", ascending=False)
                 .reset_index(drop=True)
                 .head(60)
                 [["Name", "Team",
                   "Goals", "Assists", "Ds", "Turns", "Height"]]
                 .rename(columns={"Goals": "Gs",
                                  "Assists": "As",
                                  "Turns": "Ts",
                                  })
          )
```

```python
roster.loc[(roster.Height < "5'8\"") &
           ((roster.Height < "5'10") | (roster.Height > "5'11\""))].sort_values("+/-", ascending=False)
```

```python
shorties = roster.loc[
    roster.Name.isin(["Ben Sadok", "Mark Rauls", "Nathan Kwon",
                  "Lucas Chen", "Xander Cuizon Tice", "Anders Juengst",
                    "John Miller"])].sort_values("+/-", ascending=False)[[
    "Name", "Height", "Goals", "Assists", "Ds", "Turns", "Team", "No."]].set_index("No.")
```
